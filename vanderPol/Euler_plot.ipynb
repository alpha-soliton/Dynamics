```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
def euler(x,v,a,b,c,h):
    dx = v;
    dv = v * (a - c * x *x) - b * x
    
    x = x + h * dx
    v = v + h * dv
    
    return x,v
```

```python
x = 2.
v = 1.

h = 0.01

a = 1.
b = 0.394
c = 1.

T = 50 * (1/h) #second
intT = int(T) #number of points

time = np.zeros(intT)

x_t = np.zeros(intT)
v_t = np.zeros(intT)

x_t[0] = x
v_t[0] = v
1
for i in range(intT-1):
    #print(i)
    time[i+1] = i * h
    x_t[i+1],v_t[i+1] = euler(x_t[i],v_t[i],a,b,c,h)
```

```python
#fig, ax = plt.subplots(1,1,figsize(8,4))
print(x_t)
plt.plot(time, x_t)
```

```python
from numpy.random import *
#wn = normal(0,0.5)

def euler_wn(x,v,a,b,c,h,wn):
    # white noise added
    dx = v;
    dv = v * (a - c * x *x) - b * x + wn
    
    x = x + h * dx
    v = v + h * dv
    
    return x,v

x_t_wn = np.zeros(intT)
v_t_wn = np.zeros(intT)

x_t_wn[0] = x
v_t_wn[0] = v

for i in range(intT-1):
    #wn = normal(0,0.5)
    wn = (rand()-0.5)/2
    x_t_wn[i+1],v_t_wn[i+1] = euler_wn(x_t_wn[i],v_t_wn[i],a,b,c,h,wn)
print(x_t_wn)    
plt.plot(time, x_t_wn)
```
