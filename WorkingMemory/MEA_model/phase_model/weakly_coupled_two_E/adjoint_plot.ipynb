```python
import numpy as np
import matplotlib.pyplot as plt
import math

from matplotlib.colors import BoundaryNorm
from matplotlib.ticker import MaxNLocator
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

import os
import random
import gc
import h5py

from matplotlib.backends.backend_pdf import PdfPages
```

```python
d = np.loadtxt('adjoint_20_over.dat')
```

```python
d
```

```python
adjoint = np.zeros(int(d.size/2))
time_index = np.zeros(int(d.size/2))


for i in range(int(d.size/2)):
    adjoint[i] = d[i, 1]
    time_index[i] = (d[i, 0]/5.9)*(2*math.pi)
```

```python
pp = PdfPages('adjoint.pdf')
plt.clf()
fig = plt.figure(figsize=(20,10))
plt.plot(time_index, adjoint)
#plt.scatter([0],[0], s=100, facecolor="None",edgecolors="black")
#plt.scatter([-np.pi, np.pi], [0, 0], marker="x", s=100)

plt.xlabel('φ', fontsize=24)
plt.ylabel('Z', fontsize=20)

plt.xlim(xmin=-0.02, xmax=2*np.pi*1.02)
#plt.ylim(ymin=-15.02,ymax=15.02)


ax = plt.gca()
ax.spines["right"].set_color("None")
ax.spines["left"].set_color("None")
ax.spines["top"].set_color("None")
ax.spines["bottom"].set_color("None")

xticks = np.array(["0", "π", "2π"])
locs=np.linspace(0,2*np.pi,3)
plt.xticks(locs, xticks, fontsize=18)
#plt.yticks(color="None")
#plt.vlines(np.linspace(-np.pi,np.pi,3),-8,20,linestyle=":",lw=1)
plt.vlines(np.linspace(0,2*np.pi,2),-8,25,linestyle="-",lw=1)
plt.hlines([0], 0, 2*np.pi,linestyle=":",lw=1)
plt.hlines([-8,25], 0, 2*np.pi,linestyle="-",lw=1)

plt.show()
pp.savefig(fig)
pp.close()
```
