```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def gain_function(z, alpha):
    return alpha * math.log(1 + math.exp(z/alpha))

def single_population_rate_model(E_t, u_t, x_t, J, tau, E_0, t_fac, t_d, U, dt):
    du = (U - u_t) / t_fac + U * (1 - u_t) * E_t    
    dE = (- E_t + gain_function((J * u_t * x_t * E_t + E_0), alpha)) / tau
    dx = (1 - x_t) / t_d - u_t * x_t * E_t
    
    E_t_ = E_t + dE * dt
    x_t_ = x_t + dx * dt
    u_t_ = u_t + du * dt
    return E_t_, u_t_, x_t_

E = 0.
u = 1.
x = 0.

#second
tau = 0.013
# depressing t_d > t_fac 
# facilitating t_d < t_fac    ++++   prefrontal cortex (PFC) facilitating excitatory connection
t_d = 0.2
t_fac = 1.5

alpha = 1.5

U = 0.3
E_0 = -2.
J = 4

dt = 0.0001

T = 10 * (1/dt)
intT = int(T)

time = np.zeros(intT)

E_t = np.zeros(intT)
u_t = np.zeros(intT)
x_t = np.zeros(intT)

E_t[0] = E
u_t[0] = u
x_t[0] = x

for i in range(intT - 1):
    time[i+1] = i * dt
    E_t[i+1], u_t[i+1], x_t[i+1] = single_population_rate_model(E_t[i], u_t[i], x_t[i], J, tau, E_0, t_fac, t_d, U, dt)
```

```python
plt.plot(time, E_t, label="E")
plt.plot(time, u_t, label="u")
plt.plot(time, x_t, label="x")
plt.legend(loc="lower right")
```
