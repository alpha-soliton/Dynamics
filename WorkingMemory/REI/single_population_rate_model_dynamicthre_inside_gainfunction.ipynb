```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def gain_function_(z, alpha):
    return alpha * math.log(1 + math.exp(z/alpha))

def single_population_rate_model_(E_t, u_t, x_t, z_t, J, tau, E_0, t_fac, t_d, t_th, thre_effi, U, dt):
    du = (U - u_t) / t_fac + U * (1 - u_t) * E_t    
    dE = (- E_t + gain_function_((J * u_t * x_t * E_t + E_0 - thre_effi * z_t), alpha)) / tau
    dx = (1 - x_t) / t_d - u_t * x_t * E_t
    dz = (E_t - z_t)/t_th
    
    E_t_ = E_t + dE * dt
    x_t_ = x_t + dx * dt
    u_t_ = u_t + du * dt
    z_t_ = z_t + dz * dt
    return E_t_, u_t_, x_t_, z_t_

E = 0.
u = 1.
x = 0.
z = 0.

#second
tau = 0.13
# depressing t_d > t_fac 
# facilitating t_d < t_fac    ++++   prefrontal cortex (PFC) facilitating excitatory connection
t_d = 2
t_fac = 15

t_th = 0.1
thre_effi = 0.02

alpha = 0.985

U = 0.3
E_0 = -2.3
J = 8

dt = 0.0001

T = 100 * (1/dt)
intT = int(T)

time = np.zeros(intT)

E_t = np.zeros(intT)
u_t = np.zeros(intT)
x_t = np.zeros(intT)
z_t = np.zeros(intT)

out_of_g = np.zeros(intT)
input_t = np.zeros(intT)

E_t[0] = E
u_t[0] = u
x_t[0] = x
z_t[0] = z

out_of_g[0] = gain_function_((J * u_t[0] * x_t[0] * E_t[0] + E_0 - thre_effi * z_t[0]), alpha)
input_t[0] = J * u_t[0] * x_t[0] * E_t[0] + E_0

for i in range(intT - 1):
    time[i+1] = i * dt
    E_t[i+1], u_t[i+1], x_t[i+1], z_t[i+1] = single_population_rate_model_(E_t[i], u_t[i], x_t[i], z_t[i], J, tau, E_0, t_fac, t_d, t_th, thre_effi, U, dt)
    out_of_g[i+1] = gain_function_((J * u_t[i+1] * x_t[i+1] * E_t[i+1] + E_0 - thre_effi * z_t[i+1]), alpha)
    input_t[i+1] = J * u_t[i+1] * x_t[i+1] * E_t[i+1] + E_0
```

```python
plt.clf()
fig = plt.figure(figsize = (40, 10), dpi = 600)
plt.plot(time, E_t, label="E")
plt.plot(time, u_t, label="u")
plt.plot(time, x_t, label="x")
plt.plot(time, z_t, label="threshold")
plt.legend(loc="lower right")
plt.show()
```
